```python
# Data Cleaning and Preprocessing
df_cleaned = df.copy()
```

```python
# Convert 'Reviews' to numeric
df_cleaned['Reviews'] = pd.to_numeric(df_cleaned['Reviews'], errors='coerce')
```

```python
# Convert 'Installs' to numeric by removing '+' and ','
df_cleaned['Installs'] = df_cleaned['Installs'].str.replace('+', '', regex=False).str.replace(',', '', regex=False)
df_cleaned['Installs'] = pd.to_numeric(df_cleaned['Installs'], errors='coerce')
```

```python
# Clean and preprocess the data
# Filter for paid apps only
df['Type'] = df['Type'].str.strip()  # Clean any extra whitespace
paid_apps = df[df['Type'] == 'Paid'].copy()

# Convert 'Price' and 'Installs' columns to numeric
paid_apps['Price'] = paid_apps['Price'].replace(r'[\$,]', '', regex=True).astype(float)
paid_apps['Installs'] = paid_apps['Installs'].replace(r'[\+,]', '', regex=True).astype(float)

# Calculate revenue
paid_apps['Revenue'] = paid_apps['Price'] * paid_apps['Installs']

# Remove rows with missing or zero values in relevant columns
paid_apps = paid_apps.dropna(subset=['Revenue', 'Installs', 'Category'])
paid_apps = paid_apps[(paid_apps['Revenue'] > 0) & (paid_apps['Installs'] > 0)]
```

## 1. Create a scatter plot to visualize the relationship between revenue and the number of installs for paid apps only. Add a trendline to show the correlation and color-code the points based on app categories.

```python
# Plot scatter plot with trendline
plt.figure(figsize=(15, 12))
sns.scatterplot(
    data=paid_apps,
    x='Installs',
    y='Revenue',
    hue='Category',
    palette='pastel',  
    alpha=0.7,edgecolor='r',
    s=200
)
# Add a trendline
sns.regplot(
    data=paid_apps,
    x='Installs',
    y='Revenue',
    scatter=False,
    color='green',
    line_kws={"linewidth": 2, "linestyle": "dashed"}
)

plt.title("Relationship Between Revenue and Installs for Paid Apps", fontsize=16)
plt.xlabel("Number of Installs", fontsize=15)
plt.ylabel("Revenue (USD)", fontsize=15)
legend = plt.legend(bbox_to_anchor=(1.02, 1),loc='upper left', title='Category')
plt.setp(legend.get_title(), fontsize=14, fontweight='bold', color='darkred',fontstyle='italic')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 2. Use a grouped bar chart to compare the average rating and total review count for the top 10 app categories by number of installs. Filter out any categories where the average rating is below 4.0 and size below 10 M and last update should be Jan month . this graph should work only between 3PM IST to 5 PM IST apart from that time we should not show this graph in dashboard itself.

```python
df.head()
```

```python
# Standardize 'Size' column to numeric (in MB)
def size_to_mb(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024  # Convert KB to MB
    else:
        return np.nan

df_cleaned['Size'] = df_cleaned['Size'].apply(size_to_mb)

# Convert 'Last Updated' to datetime
df_cleaned['Last Updated'] = pd.to_datetime(df_cleaned['Last Updated'], errors='coerce')

```

```python
# Filter conditions
filtered_data = df_cleaned[
    (df_cleaned['Rating'] >= 4.0) &
    (df_cleaned['Installs'] >= 10_000_000) &
    (df_cleaned['Last Updated'].dt.month == 1)
]
```

```python
# Compute metrics: Average Rating and Total Reviews for top 10 categories by installs
top_categories = (
    filtered_data.groupby('Category')
    .agg({'Rating': 'mean', 'Reviews': 'sum', 'Installs': 'sum'})
    .sort_values(by='Installs', ascending=False)
    .head(10)
)
```

```python
# Prepare data for visualization
top_categories.reset_index(inplace=True)
top_categories
```

```python
# Visualization: Grouped Bar Chart
fig, ax = plt.subplots(figsize=(15,10))

# Bar width and positions
bar_width = 0.45
categories = top_categories['Category']
x = np.arange(len(categories))

# Bars for average rating and total reviews
ratings = top_categories['Rating']
reviews = top_categories['Reviews']/1e6 # Convert reviews to millions for readability

bar1 = ax.bar(x - bar_width/3, ratings, bar_width, label='Average Rating', color='yellow')
bar2 = ax.bar(x + bar_width/3, reviews, bar_width, label='Total Reviews (Millions)', color='skyblue')

# Labels, title, and legend
ax.set_xlabel('Categories', fontsize=14)
ax.set_ylabel('Values', fontsize=14)
ax.set_title('Comparison of Average Rating and Total Reviews (Top 10 Categories)', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

## 3. Create a violin plot to visualize the distribution of ratings for each app category, but only include categories with more than 50 apps and app name should contain letter “C” and exclude apps with fewer than 10 reviews and rating should be less 4.0. this graph should work only between 4 PM IST to 6 PM IST apart from that time we should not show this graph in dashboard itself.

```python
# Step 1: Clean and filter data
df = df.dropna(subset=['Reviews', 'Rating'])

# Apply the filters
filtered_data = df_cleaned[
    (df_cleaned['Category'].map(df['Category'].value_counts()) > 50) &  # Categories with >50 apps
    (df_cleaned['App'].str.contains('C', case=False, na=False)) &                   # App names with 'C'
    (df_cleaned['Reviews'] >= 10) &                                                # Reviews >= 10
    (df_cleaned['Rating'] < 4.0)                                                   # Ratings < 4.0
]

# Step 2: Get current time (simulate real-time condition)
current_time = datetime.now().time()
start_time = datetime.strptime("16:00", "%H:%M").time()
end_time = datetime.strptime("18:00", "%H:%M").time()

# Step 3: Plot the violin graph only if within time range
if start_time <= current_time <= end_time:
    plt.figure(figsize=(15,7))
    sns.violinplot(data=filtered_data, x='Category', y='Rating', palette='bright',hue='Category')
    plt.title("Distribution of Ratings for App Categories", fontsize=16)
    plt.xlabel("App Category", fontsize=13)
    plt.ylabel("Rating", fontsize=15)
    plt.xticks(rotation=60, fontsize=10)
    
    plt.show()
else:
    print("The graph is only available between 4 PM and 6 PM IST.")
```
